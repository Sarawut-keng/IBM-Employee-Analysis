```python
def overtimewithattrition(x,y):
    oyay = len(x[x=='Yes'][y == 'Yes'])/len(x[x=='Yes'])*100
    oyan = 100 - oyay
    onay = len(x[x=='No'][y== 'Yes'])/len(x[x=='No'])*100
    onan = 100 - onay
    over1 = [oyay,oyan]
    over2 = [onay,onan]
    labels = 'Attrition = yes','Attrition = No'
    explode = [0.1,0]
    colors = ['#ff3232','#5ea998']
    fig, ax = plt.subplots(dpi = 150)
    ax.pie(over1,explode = explode, labels=labels, autopct='%1.1f%%',colors=colors,
    shadow=True, startangle=90)
    ax.set_title('Overtime = Yes')
    ax.axis('equal')
    fig2, ax2 = plt.subplots(dpi = 150)
    ax2.pie(over2,explode = explode, labels=labels, autopct='%1.1f%%',colors=colors,
    shadow=True, startangle=90)
    ax2.set_title('Overtime = No')
    ax2.axis('equal')
    plt.show()
overtimewithattrition(Overtime,Attrition)
```

```python
def pct(header,categories):
    for i in header:
        cate_pct = {}
        l = 0
        for j in categories:
            index = Data[(Data['{}'.format(i)]=='{}'.format(j))]
            yes = len(index[(index['Attrition']=='Yes')])
            no = len(index[(index['Attrition']=='No')])
            cate_pct[l] = yes/(yes+no)*100
            print("{} % of {} said Yes for attrition".format(round(cate_pct[l],2),j))
            l += 1
```

```python
headdata = ['Department']
subdata = ['Sales','Research & Development','Human Resources']
pct(headdata,subdata)
```

```python
headdata = ['Gender']
subdata = ['Male','Female']
pct(headdata,subdata)
```

```python
def pctint(header,categories):
    for i in header:
        cate_pct = {}
        l = 0
        for j in categories:
            index = Data[(Data['{}'.format(i)]==j)]
            yes = len(index[(index['Attrition']=='Yes')])
            no = len(index[(index['Attrition']=='No')])
            cate_pct[l] = yes/(yes+no)*100
            print("{} % of {} said Yes for attrition".format(round(cate_pct[l],2),j))
            l += 1
```

```python
headdata = ['StockOptionLevel']
subdata = [0,1,2,3]
pctint(headdata,subdata)
```

```python
headdata = ['PerformanceRating']
subdata = [3,4]
pctint(headdata,subdata)
```
